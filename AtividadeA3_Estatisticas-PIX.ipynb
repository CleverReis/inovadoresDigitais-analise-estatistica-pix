```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# leitura dos dados
chavePix = pd.read_csv("chavePix.csv", delimiter= ";")
transPix = pd.read_csv('transPix.csv', delimiter= ";")
```

```python
#preparação dos dados
print("Dados do arquivo chavePIX.csv")
chavePix.head()
```

```python
print("Dados do arquivo chavePIX.csv")
transPix.head()
```

```python
chavePix.info()
print()
transPix.info()
```

```python
# Remover duplicatas
chavePix.drop_duplicates(inplace=True)
transPix.drop_duplicates(inplace=True)
```

```python
#Remoção de colunas desnecessárias

chavePix.columns
```

```python
chavePix = chavePix.drop(columns=["ISPB"])
chavePix.columns
```

```python
transPix.columns
```

```python
transPix = transPix.drop(columns=['Municipio_Ibge', 'Estado_Ibge', 'Sigla_Regiao'])
transPix.columns
```

```python
chavePix.head(10)
```

```python
transPix.head(10)
```

```python
# padronizando o campo data
chavePix['coluna_data'] = pd.to_datetime(chavePix['Data'], errors='coerce')
transPix['coluna_data'] = pd.to_datetime(transPix['Data'], errors='coerce')
```

```python
chavePix['Data'] = pd.to_datetime(chavePix['Data'])

chavePix['Data'] = chavePix['Data'].dt.strftime('%m/%Y')
chavePix.head()
```

```python
transPix.rename(columns={'AnoMes': 'Data'})

transPix['Data'] = pd.to_datetime(transPix['Data'])

transPix['Data'] = transPix['Data'].dt.strftime('%m/%Y')
transPix.head()
```

```python
print(chavePix.describe())
```

```python
#não consigo juntar os arquivos
#dadosPix = pd.merge(transPix, chavePix, on='Data', how='inner')
```

```python
# Quantidade de chaves PIX por tipo de chave
qtd_chaves_por_tipo = chavePix.groupby('TipoChave')['qtdChaves'].sum().reset_index()
print("\nQuantidade de chaves PIX por tipo de chave:")
print(qtd_chaves_por_tipo)
```

```python
# Valor total transacionado por região
valor_total_por_regiao = transPix.groupby('Regiao')[['VL_PagadorPF', 'VL_PagadorPJ', 'VL_RecebedorPF', 'VL_RecebedorPJ']].sum().reset_index()
print("\nValor total transacionado por região:")
print(valor_total_por_regiao)
```

```python
# Quantidade de transações por estado
qtd_transacoes_por_estado = transPix.groupby('Estado')[['QT_PagadorPF', 'QT_PagadorPJ', 'QT_RecebedorPF', 'QT_RecebedorPJ']].sum().reset_index()
print("\nQuantidade de transações por estado:")
print(qtd_transacoes_por_estado)
```

```python
# Quantidade total de chaves por natureza do usuário
qtd_chaves_por_natureza = chavePix.groupby('NaturezaUsuario')['qtdChaves'].sum().reset_index()
print("\nQuantidade total de chaves por natureza do usuário:")
print(qtd_chaves_por_natureza)
```

```python
# Gráfico de quantidade de chaves por tipo de chave
plt.figure(figsize=(10, 6))
plt.bar(qtd_chaves_por_tipo['TipoChave'], qtd_chaves_por_tipo['qtdChaves'], color='skyblue')
plt.xlabel('Tipo de Chave')
plt.ylabel('Quantidade de Chaves')
plt.title('Quantidade de Chaves PIX por Tipo de Chave')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de valor total transacionado por região
# Limpar os dados e converter colunas de valores monetários em números
for coluna in ['VL_PagadorPF', 'VL_PagadorPJ', 'VL_RecebedorPF', 'VL_RecebedorPJ']:
    chavePix[coluna] = chavePix[coluna].str.replace('.', '').str.replace(',', '.').astype(float)

valor_total_por_regiao.set_index('Regiao')[['VL_PagadorPF', 'VL_PagadorPJ', 'VL_RecebedorPF', 'VL_RecebedorPJ']].plot(kind='bar', figsize=(12, 8))
plt.xlabel('Região')
plt.ylabel('Valor Total Transacionado')
plt.title('Valor Total Transacionado por Região')
plt.xticks(rotation=45)
plt.show()
```

```python
estados = qtd_transacoes_por_estado['Estado']
qtd_pagador_pf = qtd_transacoes_por_estado['QT_PagadorPF']
qtd_pagador_pj = qtd_transacoes_por_estado['QT_PagadorPJ']
qtd_recebedor_pf = qtd_transacoes_por_estado['QT_RecebedorPF']
qtd_recebedor_pj = qtd_transacoes_por_estado['QT_RecebedorPJ']

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(estados, qtd_pagador_pf, label='Pagador PF')
plt.bar(estados, qtd_pagador_pj, bottom=qtd_pagador_pf, label='Pagador PJ')
plt.bar(estados, qtd_recebedor_pf, bottom=qtd_pagador_pf + qtd_pagador_pj, label='Recebedor PF')
plt.bar(estados, qtd_recebedor_pj, bottom=qtd_pagador_pf + qtd_pagador_pj + qtd_recebedor_pf, label='Recebedor PJ')

# Adicionar legendas e rótulos
plt.xlabel('Estado')
plt.ylabel('Quantidade de Transações')
plt.title('Quantidade de Transações por Estado e Tipo')
plt.xticks(rotation=45)
plt.legend()

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
